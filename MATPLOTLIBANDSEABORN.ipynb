```python
import numpy as np
import matplotlib.pyplot as plt
tigar = np.linspace(-2*np.pi, 2*np.pi, 100)
print(tigar)
plt.plot(tigar, np.sin(tigar) ,color='black',linestyle='dotted')
plt.title("sin(x)")
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.plot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth = '3',label ='CompanyA',marker='*',ms='15',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label ='CompanyB',marker = 'H')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
print(iris)
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,label="inside")
plt.title("Scatter plot of sepal vs petal")
plt.xlabel("Total Bill($)")
plt.ylabel("Tip($)")
plt.legend(loc="best")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
print(iris)
sns.barplot(x="sepal_length",y="sepal_width",data=iris,label="inside")
plt.title("Scatter plot of sepal vs petal")
plt.xlabel("Total Bill($)")
plt.ylabel("Tip($)")
plt.legend(loc="best")
plt.show()
```
